```python
import pandas as pd

df = pd.read_csv(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\4\ml\titanic.csv')
print(df.head())
```

```python
# Check the shape (rows, columns)
print(df.shape)

# Get summary info about data types and non-null counts
print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
# Example: Total sales by sales representative
sales_by_rep = df.groupby('Pclass')['Survived'].sum()
print(sales_by_rep)

# Example: Total units sold by item
units_by_item = df.groupby('Age')['Fare'].sum()
print(units_by_item)
```

```python
import matplotlib.pyplot as plt

# Plot total sales by representative as a bar chart
sales_by_rep.plot(kind='bar')
plt.ylabel('Sales')
plt.show()

# Plot units sold by item as a bar chart
units_by_item.plot(kind='bar')
plt.ylabel('Units')
plt.show()
```
